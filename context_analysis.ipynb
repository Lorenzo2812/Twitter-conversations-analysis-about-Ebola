Creating a list from the column 'context_annotations', appending the string 'no context' if the observations is without a value or with the empty list as value

```python
la=[]
for i in df['context_annotations']:
    if i=='[]' or pd.isnull(i)==True:
        la.append('No context')
    else:
        x = ast.literal_eval(i)
        la.append(x)
la
```

Create a new list of lists where every list contains 'no context' for observations with no context or the names of the domains of the context.

```python
la2=[]
for i in la:
    la3=[]
    if i=='No context':
        la3.append(i)
    else:
        for j in range(len(i)):
            la3.append(i[j]['domain']['name'])
    la2.append(la3)
la2
```

Most common domains in the tweets

```python
combined_list = [domain for lists in la2 for domain in lists]
counts = Counter(combined_list)
most_common_domains = counts.most_common(30)
most_common_domains
```

Create a new list of lists where every list contains 'no context' for observations with no context or the names of the entities of the context.

```python
la22=[]
for i in la:
    la33=[]
    if i=='No context':
        la33.append(i)
    else:
        for j in range(len(i)):
            la33.append(i[j]['entity']['name'])
    la22.append(la33)
la22
```

Most common entities in the tweets

```python
combined_list2 = [entity for lists in la22 for entity in lists]
counts2 = Counter(combined_list2)
most_common_entities = counts2.most_common(30)
most_common_entities
```

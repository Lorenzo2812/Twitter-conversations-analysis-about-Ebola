# GEO-SPATIAL ANALYSIS

Geo-spatial analysis is the process of analyzing data that is geographically referenced: data that has location information associated with it. The aim is to find the locations of the tweets for which coordinates are available.

The non-null observations of the column geo are 1610

```python
dfgeo = df[df['geo'].notna()]
dfgeo
```

The Dataframe is further reduced, to obtain data containing only the observations with coordinates

```python
dfgeocor=dfgeo[dfgeo['geo'].str.contains('coordinates')]
dfgeocor = dfgeocor.reset_index(drop=True)
dfgeocor
```

```python
geo= dfgeocor['geo']
geo
```

Creation of a list of data in column geo (list of strings)

```python
lsg=[]
for i in geo:
    lsg.append(i)
lsg
```

Transformation of the previous list in a list of dictionaries

```python
lst=[]
for j in lsg:
    j = j.replace("\'", "\"")
    res = json.loads(j)
    lst.append(res)
lst
```

Coordinates are retrieved for every dictionary in the list

```python
lstt=[]
for i in lst:
    lstt.append(i['coordinates']['coordinates'])
lstt
```

A dataframe of coordinates is created from the list.

```python
dfll=pd.DataFrame(lstt,columns=['Long','Lat'])
dfll
```

We get the dataframe of the 21 observations with coordinates with a column for longitude and another one for latitude by concatenating the reduced dataframe with the dataframe with coordinates.

```python
df_geo= pd.concat([dfgeocor, dfll], axis=1)
df_geo
```

We get the country using the coordinates 

```python
lsl=[]
for i,j in zip(df_geo['Long'],df_geo['Lat']):
    geolocator = Nominatim(user_agent="geoapiExercises")
    location = geolocator.reverse(Point(j, i))
    address = location.raw['address']
    lsl.append(address['country'])
    
lsl
```

We concatenate the countries to the reduced dataframe

```python
dflsl=pd.DataFrame(lsl,columns=['Country'])
dflsl
df_geo= pd.concat([df_geo, dflsl], axis=1)
df_geo
```

We create the word map

```python
map_world = folium.Map(location=[df_geo.Lat.mean(), df_geo.Long.mean()],
                       tiles = 'openstreetmap', 
                       zoom_start = 2)
```

Circles marker related to the places identified by the coordinates are added to the map

```python
for lat, lng, label in zip(df_geo.Lat, df_geo.Long, df_geo.Country):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=f'Country: {label}',
        fill=True,
        color='black',
        fill_color='red',
        fill_opacity=0.6
        ).add_to(map_world)
map_world
```

Visualization of clusters of the markers on the map

```python
# Marker cluster
marker_cluster = MarkerCluster().add_to(map_world)
for lat, lng, label in zip(df_geo.Lat, df_geo.Long, df_geo.Country):
    folium.Marker(location=[lat,lng],
            popup = f'Location: {label}', 
            icon = folium.Icon(color='beige')
    ).add_to(marker_cluster)

    
    
map_world.add_child(marker_cluster)
```

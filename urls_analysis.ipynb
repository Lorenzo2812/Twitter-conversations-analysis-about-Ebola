# URLS ANALYSIS

The aim of this notebook is to retrieve the sources from the URLS located in the column 'original_text' of the dataframe.

Firstly, a list of lists has been created: every list contains the URLS located in a document (if there are not URLS in that document an empty list is returned)

```python
lurls=[]
for doc in df['original_text']:
    urls = re.findall(r'(https?://[^\s]+)', doc)
    lurls.append(urls)
lurls
```

There are documents containing up to ten URLS so a dataframe of URLS is created

```python
dfurl=pd.DataFrame(lurls,columns=['URL1','URL2','URL3','URL4','URL5','URL6','URL7','URL8','URL9','URL10'])
dfurl
```

This dataframe is concatenated to the original dataframe

```python
df_url= pd.concat([df, dfurl], axis=1)
```

Most common URLS in the column URL1: 11 URLS appear more than 100 times

```python
most_common = df_url['URL1'].value_counts()
most_common.nlargest(11)
```

A list of the 100 most frequent URLS is created

```python
top_100_cats = most_common.head(100)
# Convert the Series to a list
top_100_cats = top_100_cats.index.tolist()
top_100_cats
```

A list of the sources of these urls is created

```python
warnings.filterwarnings("ignore")
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'}
l_sources=[]
for url in top_100_cats:
    response = requests.get(url, verify=False, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    html_tags=soup.find('title').text
    l_sources.append(html_tags)
l_sources             
```

A list containing only the domain has been created

```python
l_agencies=[]
for url in l_sources:
    domain = get_domain_name(url)
    l_agencies.append(domain)
l_agencies   
```

From the titles we can observe that the two most frequent URLS bring to the same web page: the page change.org, where a petition is called to ask the resignation of the WHO general director Ghebreyesus for having underestimated coronavirus.

Another frequent source to which most of the tweets redirect to is mindinghearts.com : an article dealing with the esposition of the two international organization W.H.O. & U.N..

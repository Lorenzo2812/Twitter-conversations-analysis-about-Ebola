# TOPIC MODELLING

Topic modelling is an unsupervised machine learning technique that helps to uncover the topics that are present in a large corpus of text. 

In this notebook topic modelling is performed using the Latent Dirichlet Allocation (LDA) Algorithm.

Running the useful notebooks

TfidfVectorizer is used to assign a weight to each word in a document. The vectorizer is fitted on the provided data and transforms the text into a sparse matrix of token counts: a matrix of features where each row represents a document and each column represents a word with its corresponding TF-IDF score.

```python
vectorizer = TfidfVectorizer(analyzer='word',
                             max_df=0.8,
                             min_df=3,                        
                             stop_words='english',             
                             lowercase=True,                   
                             token_pattern='[a-zA-Z0-9]{3,}',  
                             max_features=5000,             
                            )

data_vectorized = vectorizer.fit_transform(corpus_preprocessed)
print(vectorizer.get_feature_names_out())
```

In this case the matrix has 93971 rows (documents) and 5000 columns (words)

```python
data_vectorized
```

Specification of LatentDirichletAllocation algorithm with 15 topics: the output (lda_ouput) is the probability of belonging to every topic for every document

```python
lda_model = LatentDirichletAllocation(n_components=15,
                                      learning_method='online',
                                      random_state=0,       
                                      n_jobs = -1  
                                     )
lda_output = lda_model.fit_transform(data_vectorized)
```

Matrix containing the probability of each word in the vocabulary belonging to each of the identified topics

```python
lda_model.components_
```

Interactive visualization of the topics with pyLDAvis

```python
warnings.filterwarnings("ignore")
pyLDAvis.enable_notebook()
pyLDAvis.sklearn.prepare(lda_model, data_vectorized, vectorizer, mds='tsne') 
```

# K-MEANS CLUSTERING

K-means Is an unsupervised machine learning algorithm that is used to group data points into clusters

Is an alternative way with respect to topic modelling with LDA for identification of topics from a corpus of text

Creation of the matrix of features where each row represents a document and each column represents a word with its corresponding TF-IDF score.

```python
stop_words = nltk.corpus.stopwords.words('english')
stop_words = stop_words
cv = TfidfVectorizer(ngram_range=(1, 2), min_df=10, max_df=0.8, stop_words=stop_words)
cv_matrix = cv.fit_transform(corpus_preprocessed)
cv_matrix.shape
```

K-means with 15 as number of clusters

```python
NUM_CLUSTERS = 15
km = KMeans(n_clusters=NUM_CLUSTERS, max_iter=10000, n_init=50, random_state=42).fit(cv_matrix)
km
```

Number of observations for every cluster

```python
Counter(km.labels_)
```

A column with cluster membership for each element of the dataframe is created

```python
df['kmeans_cluster'] = km.labels_
df
```

Key features and popular tweets for each cluster:

```python
feature_names = cv.get_feature_names()
topn_features = 15
ordered_centroids = km.cluster_centers_.argsort()[:, ::-1]

for cluster_num in range(NUM_CLUSTERS):
    key_features = [feature_names[index] 
                        for index in ordered_centroids[cluster_num, :topn_features]]
    tweets = tw_clusters[tw_clusters['kmeans_cluster'] == cluster_num]['original_text'].values.tolist()
    print('CLUSTER #'+str(cluster_num+1))
    print('Key Features:', key_features)
    print('Popular Tweets:', tweets)
    print('-'*80)
```

Creation of a list of dictionaries in which in every dictionary there are the words for that theme as keys and for each key the number of tweets containing that word as value 

```python
word_count = [{word: 0 for word in total_list[i]} for i in range(len(total_list))]
for ls in total_list:
    for word in ls:
        for doc in df['unigrams']:
            for unigram in doc:
                if word==unigram:
                    word_count[total_list.index(ls)][word] += 1
print(word_count)
```
